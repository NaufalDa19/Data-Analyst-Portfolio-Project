```python
# ============================================================================
# PANDUAN MEMILIH METODE HANDLING MISSING DATA
# ============================================================================
# 
# 1. Missing < 5% dari total data:
#    → DROP rows dengan dropna()
# 
# 2. Missing 5-40%:
#    Untuk data NUMERIK:
#    → Fill dengan MEDIAN (lebih robust terhadap outliers)
#    → Fill dengan MEAN (jika distribusi data normal)
#    → KNN Imputation (jika ada korelasi antar features)
#    
#    Untuk data CATEGORICAL:
#    → Fill dengan MODE (nilai paling sering muncul)
#    → Fill dengan 'Unknown' atau 'Missing'
#    → Buat kategori terpisah untuk missing
# 
# 3. Missing > 40%:
#    → DROP kolom (terlalu banyak missing)
#    → ATAU gunakan advanced method (MICE, KNN) jika kolom sangat penting
# 
# 4. Time Series Data:
#    → Forward Fill / Backward Fill
#    → Interpolation (linear atau time-based)
# 
# 5. Data dengan Outliers:
#    → Gunakan MEDIAN (bukan mean)
#    → Gunakan MODE untuk categorical
# 
# 6. Feature Engineering:
#    → Buat kolom flag untuk missing indicator
#    → Missing pattern bisa jadi informasi penting
# ============================================================================
```

# Finding Correlation (Menemukan korelasi dari tiap kolom) 

```python
# Tujuan Analisis
# Analisis ini dilakukan untuk mengidentifikasi hubungan (korelasi) antar variabel
# dalam dataset, khususnya untuk mengetahui pasangan kolom yang memiliki korelasi tinggi.

# Metode Analisis
# Korelasi dianalisis dengan membandingkan nilai antar kolom numerik
# serta mengamati hubungan antara variabel independen dan variabel target (Gross).

# Hasil Analisis
# Berdasarkan hasil analisis korelasi, terdapat beberapa pasangan kolom
# yang menunjukkan potensi korelasi tinggi, yaitu:
# 1. Budget dengan Gross
# 2. Company dengan Gross

# Kesimpulan
# Budget memiliki hubungan yang cukup kuat dengan Gross, yang menunjukkan bahwa
# semakin besar anggaran produksi, semakin besar potensi pendapatan film.
# Selain itu, Company juga berpotensi memengaruhi Gross, yang mengindikasikan bahwa
# perusahaan produksi tertentu cenderung menghasilkan pendapatan yang lebih tinggi.
```

```python
# Terdapat beberapa metode yang dapat digunakan untuk mengukur korelasi, yaitu:
# 1. Pearson
# 2. Kendall
# 3. Spearman
```

```python
# Menghitung dan menampilkan nilai korelasi antar kolom numerik
df.corr(numeric_only=True)
```

```python
# Metode dalam melakukan perhitungan korelasi

# Metode Pearson digunakan untuk mengukur korelasi linear antara dua variabel numerik
# yang memiliki hubungan linier dan berdistribusi normal serta minim outlier.

# Metode Spearman digunakan untuk mengukur korelasi berdasarkan peringkat (rank),
# cocok untuk data yang tidak berdistribusi normal atau memiliki hubungan monoton
# (naik/turun tetapi tidak harus linier).

# Metode Kendall digunakan untuk mengukur kekuatan hubungan berdasarkan kesesuaian
# pasangan data (concordant dan discordant), cocok untuk dataset berukuran kecil
# dan data ordinal.
```

```python
# Melihat korelasi menggunakan metode pearson
df.corr(method = 'pearson', numeric_only = True)
```

```python
# Melihat korelasi menggunakan metode kendall
df.corr(method = 'kendall', numeric_only = True)
```

```python
# Melihat korelasi menggunakan metode spearman
df.corr(method = 'spearman', numeric_only = True)
```

```python
# Interpretasi nilai korelasi:
# Nilai korelasi berada pada rentang -1 hingga 1.
# Semakin mendekati 1 atau -1, maka hubungan antar variabel semakin kuat.

# Kategori kekuatan korelasi:
# 0.00 – 0.19  : Korelasi sangat lemah
# 0.20 – 0.39  : Korelasi lemah
# 0.40 – 0.59  : Korelasi sedang
# 0.60 – 0.79  : Korelasi kuat
# 0.80 – 1.00  : Korelasi sangat kuat

# Tanda positif (+) menunjukkan hubungan searah,
# sedangkan tanda negatif (-) menunjukkan hubungan berlawanan arah.
```

```python
# Berdasarkan hasil matriks korelasi, variabel Budget dan Gross
# menunjukkan korelasi positif yang kuat. Hal ini mengindikasikan bahwa
# semakin besar budget yang dikeluarkan, semakin besar pula potensi gross
# yang dihasilkan.
```

```python
# Karena perhitungan korelasi dilakukan pada kolom numerik dan bertujuan
# untuk mengukur hubungan linier antar variabel, maka metode korelasi
# yang digunakan adalah Pearson.
```

```python
# Menghitung matriks korelasi Pearson untuk fitur numerik
correlation_matrix = df.corr(numeric_only = True, method = 'pearson')

# Visualisasi heatmap matriks korelasi
sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Membuat salinan dataframe untuk proses numerisasi kolom kategorikal
# Kolom kategorikal akan dikonversi menjadi numeric menggunakan category codes
df_numerized = df.copy()

df_numerized.head()
```

```python
# Label encoding: mengubah setiap kolom bertipe object menjadi numerik
# Setiap nilai unik akan diberi kode angka berurutan (0, 1, 2, ...)
for col_name in df_numerized.columns:
    if (df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes

df_numerized.head()
```

```python
# Menampilkan dataframe original sebagai pembanding
df.head()
```

```python
# Menghitung matriks korelasi Pearson dari dataframe yang sudah diencode
correlation_matrix_new = df_numerized.corr(method = 'pearson')

# Visualisasi heatmap matriks korelasi
sns.heatmap(correlation_matrix_new, annot = True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
# Menampilkan matriks korelasi dalam bentuk tabel
df_numerized.corr()
```

```python
# Menghitung matriks korelasi
correlation_mat = df_numerized.corr()

# Flatten matriks korelasi menjadi pasangan fitur (feature pairs)
corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
# Mengurutkan pasangan korelasi dari nilai terendah ke tertinggi
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
# Memfilter pasangan dengan korelasi tinggi (> 0.5)
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr
```

## 1. Perjalanan Analisis: Menemukan Faktor Kunci Kesuksesan Film

Ketika pertama kali melihat dataset ini, saya memiliki hipotesis awal yang cukup meyakinkan: budget dan company adalah faktor utama yang menentukan gross earning (pendapatan kotor) sebuah film. Pemikiran ini terdengar logis—studio besar dengan anggaran besar seharusnya menghasilkan film yang menguntungkan, bukan?

Namun, data bercerita hal yang berbeda.

Setelah melalui proses Exploratory Data Analysis (EDA) yang mendalam dan pembersihan data yang teliti, saya menemukan sesuatu yang menarik. Melalui analisis korelasi terhadap setiap kolom dalam dataset, terungkap bahwa faktor yang sebenarnya memiliki korelasi tinggi dengan gross earning bukanlah company seperti yang saya duga awalnya.

Hasil analisis menunjukkan bahwa dua kolom yang memiliki korelasi paling kuat dengan gross earning adalah:
- **Budget** - Sesuai prediksi awal, anggaran produksi memang berpengaruh signifikan
- **Votes** - Ini yang mengejutkan! Jumlah votes (kemungkinan rating dari penonton) ternyata menjadi indikator kuat kesuksesan finansial film

Temuan ini memberikan insight berharga: bukan hanya soal seberapa besar budget atau seberapa terkenal studio produksinya, tetapi bagaimana film tersebut diterima oleh audiens (tercermin dari jumlah votes) yang ternyata menjadi kunci kesuksesan finansial sebuah film.

## 2. Kesimpulan

Dari analisis dataset movies ini, dapat disimpulkan bahwa terdapat dua faktor utama yang memiliki pengaruh signifikan terhadap gross earning:

- ### Budget Produksi
      Terdapat hubungan positif yang kuat antara budget dan gross earning. Semakin besar anggaran yang diinvestasikan dalam produksi film, semakin besar pula potensi pendapatan yang dapat dihasilkan. Hal ini menunjukkan bahwa investasi produksi yang memadai berperan penting dalam kesuksesan finansial sebuah film.

- ### Votes (Dukungan Audiens)
      Jumlah votes mencerminkan tingkat engagement dan popularitas film di kalangan penonton. Semakin tinggi jumlah votes, semakin besar pula gross earning yang dihasilkan. Votes menjadi indikator kuat bahwa film tersebut diminati dan ditonton oleh banyak orang, yang pada akhirnya berdampak langsung pada performa box office.

      Kedua faktor ini saling melengkapi: budget yang besar memungkinkan produksi berkualitas tinggi, sementara respon positif dari audiens (votes) mendorong lebih banyak penonton untuk menonton film tersebut, menciptakan siklus yang menguntungkan bagi gross earning.
